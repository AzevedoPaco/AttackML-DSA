This file is used to launch the attack described in the article ‘Breaking ML-DSA with an explicit hint’. More specifically, it is used to test three variations of the attacks described:

-In the first cell, the entire vector X is reconstructed from a single ML-DSA signature whose last $X$ coefficients are zero, for different values of $X$.

-In the second cell, X is reconstructed from two signatures, both with the last $X$ coefficients set to 0, again for different values of $X$.

-In the last cell, only the last coefficient of each signature is zero, the attack is ran by varying the number of known signatures.

```python
d=20
e=30
bkz_size=None

Time=[]
Success=[]
for _ in range(10):
    print("-----"+str(_)+"-----")
    attack=attack_single_sig(d,e,bkz_size,complete_atk=True)
    Time.append(float(attack[0]))
    Success.append(attack[1])
    print("")
    
print("Results: \n")
print("Average Attack Time:", sum(Time)/10.0)
print("Success probability:", sum(Success)/10.0)
```
